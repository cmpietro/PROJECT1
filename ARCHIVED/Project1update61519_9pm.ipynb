```python
response = requests.get(url).json()
#print(json.dumps(response, indent = 4, sort_keys=True))
```

```python
healthdata = pd.DataFrame(response)
healthdata.head()
```

```python
#Collecting a list of all columns within the DataFrame
healthdata.columns
```

```python
# Reorganizing the columns using double brackets
organized_df = healthdata[["year", "locationabbr", "topicdesc", "response", "data_value", "gender", "race","education", "stratificationid1", "stratificationid2", "stratificationid3", "stratificationid4"]]
organized_df.head()
```

```python
renamed_df = organized_df.rename(columns={"locationabbr":"State", "topicdesc":"Use Type", "response":"Use Frequency", "data_value":"% of Respondents"})
renamed_df.head()
```
